```python
import  numpy as np
import pandas as pd
```

```python
titanic_passangers = pd.read_csv('titanic_data.csv',index_col='PassengerId')
titanic_surv = pd.read_csv('titanic_surv.csv',index_col='PassengerId')
titanic_passangers = titanic_passangers.join(titanic_surv, lsuffix='_pasengers',rsuffix='_surv')
titanic_passangers.head()
```

```python
sex=titanic_passengers['Sex']

sex=(sex == 'male').astype(int)
titanic_passengers['Sex']=sex

titanic_passengers.head()
```

```python
titanic_passengers['Embarked'].value_counts()
```

```python
embarked=titanic_passengers['Embarked']

embarked_dict={
    'S':0,
    'C':1,
    'Q':2
}

embarked=embarked.map(embarked_dict)
titanic_passengers['Embarked']=embarked
titanic_passengers.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(titanic_passengers['Age'], titanic_passengers['Fare'], c=titanic_passengers['Survived_surv'])

plt.xlabel('Age')
plt.ylabel('Fare')
```
