```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('Restaurant.csv')
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.hist(df['total_bill'])
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.hist(df['total_bill'], bins=20, color='pink')
axes.set_title('Общая сумма счета')
axes.set_ylabel('Количество заказов')
axes.set_xlabel('Сумма заказа, USD')
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.scatter(df['total_bill'], y=df['tip'], marker='*')
axes.set_title('Общая сумма счета VS сумма чаевых')
axes.set_ylabel('сумма чека')
axes.set_xlabel('Размер чаевых')
```
