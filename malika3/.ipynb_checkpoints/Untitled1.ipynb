```python
import numpy as np

rating=np.array([
    [1, 280, 892, 621],
    [2, 254, 742, 763],
    [3, 352, 820, 757],
    [4, 189, 730, 614],
    [5, 209, 758, 608],
    [6, 312, 652, 545],
    [7, 199, 680, 553],
    [8, 224, 592, 485],
    [9, 293, 603, 466],
    [10, 187, 507, 423]
])
```

```python
mean=np.mean(rating[:, -1])
mean
```

```python
median=np.median(rating[:, -1])
median
```

```python
mean_col2=np.mean(rating[:, 1])
mean_col2
```

```python
std_col2=np.std(rating[:, 1])
std_col2
```

```python
corr=np.corrcoef(rating[:, 1], rating[:, 2])
corr
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df=pd.read_csv('Restaurant-checkpoint.csv')

tips=df[['total_bill', 'tip', 'size']]

correlation=tips.corr()

sns.heatmap(correlation, annot=True, cmap='coolwarm')
```

```python
sns.boxplot(x='day', y='tip', data=df)
```
