```python
import pandas as pd

# Load the data
football = pd.read_csv('football.csv')

# Calculate the average age and round it
age = round(football['Age'].mean())

# Describe object columns
object_description = football.describe(include=['object'])

# Fix the reference to the top nationality
# Use `football['Nationality']` for this operation
nationality_top = object_description['Nationality']['top']

# Print results
print(age, nationality_top)
```

```python
import pandas as pd

football = pd.read_csv('football.csv')

speedL = football[football.Wage < football.Wage.mean()]['SprintSpeed'].mean()
speedL = round(speedL)

age = football[(football.Composure > football.Composure.max() * 0.9) &
               (football.Reactions > football.Reactions.max() * 0.9)]

n = football[football.Wage > football.Wage.mean()]['Nationality'].describe()['top']

print(speedL, age, n)
```
